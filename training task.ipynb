Попробуем построить модель множественно линейной регрессии. За таргет значение возьмем минимальное значение однородной группы, назовем его "Оклад". Формула регрессии поможет вывести формулу оклада, высчитать коэффициенты и влияющие на оклад показатели.

Итак формула множественно регрессии: y = k1x1 + k2x2 + ... + knxn + b


Для расчета окалада, дабы избежать повторений, выведем новую таблицу. Таблица будет состоять из сотрудников, которые соответвуют уникальным группам(комбинациям) параметров и минимальному значению Дохода внутри этой группы.


```python
dfdf = df1[(df1.fio == 'Александров А. М.') | (df1.fio == 'Григорьев А. Т.') | (df1.fio == 'Кузнецов О. А.') | (df1.fio == 'Агафонов С. И.')
   | (df1.fio == 'Белкин В. И.') | (df1.fio == 'Дьяконов М. Т.') | (df1.fio == 'Власов С. Д.') | (df1.fio == 'Антонов В. Т.')
   | (df1.fio == 'Ильин Э. П.') | (df1.fio == 'Волошин Л. И.') | (df1.fio == 'Аксенов Е. Н.') | (df1.fio == 'Алексеев Н. Н.') 
    | (df1.fio == 'Высоцкий М. А.') | (df1.fio == 'Соколов И. А.') | (df1.fio == 'Виноградов П. Д.') | (df1.fio == 'Герасимов К. Я.') 
    | (df1.fio == 'Орлов Т. Д.') | (df1.fio == 'Быков Л. Н.') | (df1.fio == 'Ермаков Г. И.') | (df1.fio == 'Белов А. М.') 
    | (df1.fio == 'Егоров Г. И.') | (df1.fio == 'Демидов М. Д.')]
len(dfdf)
```

```python
#Исключим из дохода вредность, создадим новый столбец с доходом(Окладом).
dfdf['c_income'] = dfdf['income'] - (dfdf.harmfulness*180)
```

Применим метод OneHotEncoding.

```python
dfdf_new = pd.get_dummies(dfdf, columns = ["profession","equipment",'category' ],
                             prefix=["profession_type","equipment_type", 'category_type'])
dfdf_new.columns
```

```python
#взглянем на матрицу корреляций
sns.heatmap(dfdf_new.corr(),annot=True,cmap='RdYlGn',linewidths=0.2) #data.corr()-->correlation matrix
fig=plt.gcf()
fig.set_size_inches(15,15)
plt.show()
```

Т.к. Формула должна выводить доход с точностью до рубля, решаем, что не стоит убирать какие-то еще параметры из модели.

```python
#строим модель линейной регрессии
category3 = dfdf_new['category_type_3']
category4 = dfdf_new['category_type_4']
category5 = dfdf_new['category_type_5']
category6 = dfdf_new['category_type_6']
category7 = dfdf_new['category_type_7']
prof_valt = dfdf_new['profession_type_Вальцовщик']
prof_kont = dfdf_new['profession_type_Контролер качества готовой продукции']
prof_nagr = dfdf_new['profession_type_Нагревальщик металла']
prof_ogne = dfdf_new['profession_type_Огнеупорщик']
prof_oper = dfdf_new['profession_type_Оператор поста управления']
prof_podr = dfdf_new['profession_type_Подручный вальцовщика']
prof_posa = dfdf_new['profession_type_Посадчик металла']
prof_rezc = dfdf_new['profession_type_Резчик горячего металла']
prof_ubor = dfdf_new['profession_type_Уборщик']
equi_vspo = dfdf_new['equipment_type_Вспомогательные помещения']
equi_kali = dfdf_new['equipment_type_Калибровочный стан']
equi_kaligrp = dfdf_new['equipment_type_Калибровочный стан, ГРП']
equi_kont = dfdf_new['equipment_type_Контрольно-измерительная аппаратура']
equi_nagr = dfdf_new['equipment_type_Нагревательные печи']
equi_nagrkant = dfdf_new['equipment_type_Нагревательные печи, кантовка']
equi_nagrnagr = dfdf_new['equipment_type_Нагревательные печи, нагрев']
equi_pili = dfdf_new['equipment_type_Пилигримовые станы']
equi_pros = dfdf_new['equipment_type_Прошивной стан']
y = dfdf_new.c_income

x = np.vstack([np.array(category3), np.array(category4), np.array(category5),
               np.array(category6), np.array(category7),
               np.array(prof_valt), 
               np.array(prof_kont), np.array(prof_nagr), np.array(prof_ogne), 
               np.array(prof_oper), np.array(prof_podr), np.array(prof_posa), 
               np.array(prof_rezc), np.array(prof_ubor), np.array(equi_vspo), 
               np.array(equi_kali), np.array(equi_kaligrp), np.array(equi_kont), 
               np.array(equi_nagr), np.array(equi_nagrkant), np.array(equi_nagrnagr), 
               np.array(equi_pili), np.array(equi_pros)]).T

model = LinearRegression().fit(x, y)
print('intercept:', model.intercept_)

print('slope:', model.coef_)
```

Модель дала коэффициенты уравнения регрессии y = k1x1 + k2x2 + ... + knxn + b. Видно, что все коэффициенты кроме категорий очень малы. В нашем случае, применив OneHotEncoding любой 'x' в уравнении равен 1. Умножив 1 на число с подписью е-12, мы получим ничтожно маленькое значение приближенное к нулю. Соответсвенно делаем вывод, что на доход без учета вредности(c_income) влияет только категория.

Итак для  3, 4, 5, 6, 7, разряда в формулу подставляются коэффициенты: -4576, -2626, -366, 2264, 5304 соответственно. Интерсепт 16756.

Формула Оклада = 16756 + 1*коэффициент разряда.

Проверим модель:


```python
print('Предсказание модели: ', model.predict(x[0:22]))
```
